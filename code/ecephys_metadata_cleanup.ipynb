```python
from utils import create_co_client, create_doc_db_client, get_doc_db_records
import pandas as pd
```

```python
docdb_client = create_doc_db_client()

query = {
    '$and': [
        {
            'name': {
                '$regex': 'ecephys', 
                '$options': 'i'
            }
        }, {
            'procedures': None
        }, {
            'data_description': {'$ne': None}
        }
    ]
}

projection = {
    "name": 1,
    "created": 1,
    "location": 1,
    "procedures": 1,
    "metadata": 1,
    "data_description": 1,
}

records = get_doc_db_records(docdb_client, projection, query)
```

```python
print(records[0])
```

```python
issues_df = pd.DataFrame(columns=["asset_name", "investigators", "project"])

for item in records:
    
    new_row = {
        "asset_name": item['name'],
        "investigators": item.get("data_description").get("investigators"),
        "project": item.get("data_description").get("project_name")
    }
    if any(word in new_row['asset_name'] for word in ['curated', 'derived']):
        continue
    issues_df = issues_df.append(new_row, ignore_index = True)
    
issues_df.to_csv('../results/ecephys_missing_procedures.csv')
    
```

```python
query = {
    '$and': [
        {
            'name': {
                '$regex': 'ecephys', 
                '$options': 'i'
            }
        }, {
            'procedures': {
                '$ne': None
            }
        }, {
            'procedures.subject_procedures':
                '$ne': None
            }
        }, {
            'procedures.subject_procedures.procedures.injection_material': None
        }
    ]
}

projection = {
    "name": 1,
    "created": 1,
    "location": 1,
    "procedures": 1,
    "metadata": 1,
    "data_description": 1,
}

records = get_doc_db_records(docdb_client, projection, query)
print(len(records))
```

```python
issues_df = pd.DataFrame(columns=["asset_name", "procedure_type", "investigators", "experimenter", "project"])

for item in records:
    if item['procedures'].get('subject_procedures'):
        for subj_procedure in item['procedures']['subject_procedures']:
            if subj_procedure.get('procedures'):
                for procedure in subj_procedure['procedures']:
                    if any(word in procedure.get('procedure_type') for word in ['Nanoject', 'Retro']) and not procedure.get("injection_materials"):
                        
                        if item.get("data_description"):
                            investigators = item.get("data_description").get("investigators")
                            project = item.get("data_description").get("project_name")
                        else:
                            investigators = None
                            project = 'NO DATA DESC'
                        new_row_dict = {
                            "asset_name": item['name'],
                            "procedure_type": procedure.get('procedure_type'),
                            "investigators": investigators,
                            "experimenter": subj_procedure.get("experimenter_full_name"),
                            "project": project
                        }
                        issues_df = issues_df.append(new_row_dict, ignore_index = True)
    
issues_df.to_csv('../results/ecephys_missing_inj_materials.csv')        
```
